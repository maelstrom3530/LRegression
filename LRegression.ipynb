```python
import ipywidgets as widgets
from IPython.display import display

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

```

```python
data = pd.read_csv('weight-height.csv')
X = data.drop(columns=['Gender', 'Weight']) # X holds height
y = data.drop(columns=['Gender', 'Height']) # Y holds weight
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

regressor = LinearRegression()
regressor.fit(X_train,y_train)
r2_score = regressor.score(X_test,y_test)
print(r2_score*100,'%')
```

```python
linreg = LinearRegression()
```

```python
linreg.fit(X, y)
```

```python
y_predict = linreg.predict(X)
```

```python
plt.scatter(X, y, alpha=0.2)
plt.plot(X, y_predict, color='red')
plt.show()
```

```python
print(linreg.coef_)
```

```python
print(linreg.intercept_)
# y =(7.72x - 351)
```

```python
predict_button = widgets.Button(description="Click Me!")
output = widgets.Output()
display(predict_button, output)
    
slider = widgets.IntSlider(min=52, max=90, step=1, value=48, description="Set height")
display(slider)
cust_h = slider.value

def on_button_clicked(b):
    with output:
        cust_h = slider.value
        cust_w = (7.72 * cust_h) - 351
        print("For the given height of " + str(cust_h) + ", the predicted weight is " + str(cust_w) + ".")


predict_button.on_click(on_button_clicked)

```
